
# 🔍 비지도학습 실습: 붓꽃 품종 군집화

 **비지도학습(unsupervised learning)**의 대표적인 알고리즘인  
**K-Means 군집화**를 직접 실습해봅시다.

<br/>


### 실습의 목적
- 라벨이 없는 데이터를 군집화(클러스터링)하는 방법을 배웁니다.
- 시각화를 통해 군집의 형태를 쉽게 이해해봅니다.



---


<br/><br/><br/>

## 1. 데이터 불러오기 및 라이브러리 임포트
- 붓꽃(iris) 데이터셋을 불러오고, Pandas DataFrame으로 변환하는 단계입니다.
- 붓꽃(iris) 데이터셋은 꽃잎과 꽃받침의 길이/너비로 이루어진 수치형 데이터이며, 실제로는 라벨이 존재하지만 이번 실습에서는 라벨 없이 비지도학습으로 클러스터링을 진행합니다.





```python
# 여기에 코드를 작성하세요!

# 1. 라이브러리 import
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# 2. 붓꽃 데이터 불러오기
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=['sepal_length','sepal_width','petal_length','petal_width'])
iris_df.head()
```

```python
## 모범답안

# 필요한 라이브러리 import
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 데이터 불러오기
iris = load_iris()
irisDF = pd.DataFrame(data=iris.data, columns=['sepal_length','sepal_width','petal_length','petal_width'])
irisDF.head()
```

<br/><br/>

---

<br/>


## 2. KMeans 군집화 수행
- 클러스터 개수를 3으로 설정하여 군집화를 수행합니다.
- KMeans는 랜덤하게 중심점을 선택한 후, 반복적으로 군집을 조정하며 최적화합니다.

```python
# 여기에 코드를 작성하세요!
kmeans1 = KMeans(n_clusters=3, random_state=777,max_iter=10000)
kmeans1.fit(iris_df)
```

```python
## 모범답안

kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=777)
kmeans.fit(irisDF)
```

<br/><br/>

---

<br/>


## 3. 군집화 결과 확인
- 각 데이터 포인트가 어느 군집(cluster)에 속했는지 확인합니다.

```python
# 여기에 코드를 작성하세요!
print(kmeans1.labels_)
```

```python
## 모범답안

print(kmeans.labels_)
```

<br/><br/>

---

<br/><br/>



## 4. 원래 타겟값과 군집 결과 비교
- 실제 타겟(품종)과 군집 결과를 비교하여 군집화가 어느 정도 의미 있는지를 확인합니다.

```python
# 여기에 코드를 작성하세요!
iris_df['target'] = iris.target
iris_df['cluster'] = kmeans.labels_
iris_df.groupby(['target', 'cluster'])['sepal_length'].count()
```

```python
## 모범답안

irisDF['target'] = iris.target
irisDF['cluster'] = kmeans.labels_
irisDF.groupby(['target', 'cluster'])['sepal_length'].count()
```

<br/><br/>

---

<br/>


## 5. 군집 시각화


- 군집화가 잘 되었는지를 확인하려면 데이터를 2차원 공간에 시각화하는 것이 좋습니다.
- 꽃잎의 길이와 너비 같은 두 개의 특징만 선택해도 됩니다.
- petal_length와 petal_width 두 개의 특성만 사용하여 군집 결과를 시각화합니다.



```python
# 여기에 코드를 작성하세요!
plt.figure(figsize=(6,4))
plt.scatter(iris_df['petal_length'], iris_df['petal_width'], c=iris_df['cluster'], cmap='viridis') # c = 군집의 중심(center / 'cluster')
plt.colorbar(label='Cluster')
plt.show()
```

```python
## 모범답안

plt.figure(figsize=(6, 4))
plt.scatter(irisDF['petal_length'], irisDF['petal_width'], c=irisDF['cluster'], cmap='viridis')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('KMeans Clustering ')
plt.colorbar(label='Cluster')
plt.show()
```
