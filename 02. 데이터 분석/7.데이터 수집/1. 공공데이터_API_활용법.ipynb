```python
병원리스트 = result["data"]
병원리스트
```

딕셔너리로 된 데이터를 한 눈에 확인할 수 있도록 데이터프레임 `df`로 바꿔보세요.

```python
# 여기에 코드를 작성하세요!
df = pd.concat([pd.DataFrame([hospital]) for hospital in hospital_list], ignore_index=True)
display(df)
```

```python
df = pd.concat([pd.DataFrame([병원]) for 병원 in 병원리스트], ignore_index=True)
df
```
