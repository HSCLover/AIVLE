# 단변량 분석 (Univariate Analysis)
단변량 분석은 **변수 하나**만을 대상으로 분포, 중심 경향성, 이상치 등을 분석하는 기초적인 탐색적 데이터 분석 방법입니다.
<br/>

#### ✅ 분석 대상
- 숫자형 변수: 평균, 중앙값, 분산 등 통계량과 시각화
- 범주형 변수: 빈도수, 비율 및 막대그래프 등을 활용

<br/>
<br/>

## 단변량 분석 1: 숫자형 변수 분석하기



```python
import pandas as pd

# 예제 데이터 (학생 시험 점수)
data = {
    'math_score': [78, 85, 90, 95, 88, 70, 65, 60, 55, 98, 100, 74, 82, 79, 92],
}
df = pd.DataFrame(data)
df.head()
```

### 다음 문제를 해결해봅시다.


### 실습 1. 박스플롯으로 단변량 분석
수학점수를 **박스플롯(Boxplot)**으로 시각화하여

1) **중앙값**이 어디쯤 있는지,

2) 점수들이 얼마나 퍼져 있는지**(사분위수)**,

3) **이상치**가 있는지를 확인해보세요.

```python
# 여기에 코드를 작성하세요!
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='math_score', data=df, color="white")
plt.show()
```

```python
# 모범 답안
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='math_score', data=df)
plt.title('Boxplot of Math Score')
plt.xlabel('Score')
plt.show()
```

##### [결과 해석]
- 중앙값은 약 80점 근처에 위치
- 점수 분포는 상대적으로 좌우 균형 잡힌 편

- 이상치는 100점에 가까운 일부 고득점자로 간주될 수 있음 (표시된 점 유무로 확인 가능)

<br/><br/><br/>

### 실습 2. 히스토그램으로 단변량 분석
수학 점수들을 **히스토그램**으로 시각화하여

1) 점수가 어떤 범위에 많이 분포되어 있는지,

2) 왼쪽이나 오른쪽으로 치우쳐 있는지,

3) 대칭적인지 또는 한쪽으로 쏠렸는지를 확인해보세요.

히스토그램은 **빈도(개수)**를 기준으로 막대를 세워, 전체 데이터의 분포를 한눈에 보여줍니다.


```python
# 여기에 코드를 작성하세요!
plt.hist(df['math_score'],bins=8,edgecolor="black",color="skyblue")
plt.grid(True)
plt.show()
```

```python
# 모범 답안

import matplotlib.pyplot as plt


plt.hist(df['math_score'], bins=6, edgecolor='black', color='skyblue')
plt.title('Histogram of Math Score')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

##### [결과 해석]
- 점수 분포는 비교적 정규분포에 가까운 형태이나, 고득점 쪽으로 약간 치우쳐 있음
- 가장 많은 학생들이 70~90점 사이에 분포
- 점수 간격(bins)을 조절하여 세부 해석 가능




<br/><br/><br/>

### 실습 3. 밀도함수 그래프로 단변량 분석
수학점수를 **커널 밀도함수(KED) 그래프**로 시각화해 보세요.
이 그래프는 히스토그램보다 부드러운 곡선 형태로 데이터의 분포를 보여줍니다.

1) 데이터가 어디에 몰려 있는지,

2) 분포가 넓게 퍼져 있는지 등을 시각적으로 알 수 있습니다.


```python
# 여기에 코드를 작성하세요!

sns.kdeplot(df['math_score'])
plt.grid(True)
plt.show()
```

```python
# 모범 답안

import matplotlib.pyplot as plt
import seaborn as sns


sns.kdeplot(df['math_score'], shade=True, color='green')
plt.title('Density Plot of Math Score')
plt.xlabel('Score')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

##### [결과 해석]
- 곡선의 중심은 약 80점 부근에 위치
- 분포가 부드럽게 퍼져 있으며, 정규분포 형태와 유사
- 이상치(100점)는 그래프의 꼬리 부분으로 나타남



<br/><br/><br/>

## 단변량 분석 2: 범주형 변수 분석하기
<br/>



```python
import pandas as pd

# 예제 데이터: 고객 성별 및 구매 여부
data = {
    'gender': ['Male', 'Female', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'purchased': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)
df.head()
```

### 다음 문제를 해결해봅시다.


### 실습 1. 막대그래프로 단변량 분석
**gender** 변수에는 **'Male'**, **'Female'**과 같은 값들이 들어 있습니다.

이 변수의 각 값이 데이터에서 몇 번 나오는지 막대그래프로 나타내보세요.

시각화를 통해 어떤 성별이 더 많이 등장하는지 쉽게 알 수 있습니다.

```python
# 여기에 코드를 작성하세요!

sns.countplot(x='gender', data=df, palette="pastel", hue='purchased', hue_order=['Yes',"No"])
plt.grid(axis='y')
plt.show()
```

```python
# 모범 답안

import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='gender', data=df, palette='pastel')
plt.title('Count of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

##### [결과 해석]
- 전체 고객 중 여성(Female)이 남성보다 소폭 더 많음

- 단순 빈도 비교를 통해 성별 분포를 직관적으로 확인할 수 있음



<br/><br/><br/>

### 실습 2. 모자이크 플롯으로 단변량 분석
**gender**과 **purchased** 두 범주형 변수는 '남성/여성', '구매함/구매 안 함'과 같은 값을 가지고 있습니다.

이 두 변수 사이에 어떤 관련성이 있는지 **모자이크 플롯**을 통해 시각적으로 살펴보세요.

각 사각형의 크기를 비교하면 어떤 조합이 많이 나타났는지 쉽게 알 수 있습니다.

```python
# 여기에 코드를 작성하세요!
from statsmodels.graphics.mosaicplot import mosaic

crosstab = pd.crosstab(df['gender'], df['purchased'])
plt.figure(figsize=(6,5))
mosaic(crosstab.stack())
plt.show()
```

```python
# 모범 답안

from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt

cross_tab = pd.crosstab(df['gender'], df['purchased'])
plt.figure(figsize=(6, 5))
mosaic(cross_tab.stack())
plt.title("Mosaic Plot of Gender vs Purchased")
plt.show()
```

##### [결과 해석]

- 여성은 구매(Yes) 비율이 높고, 남성은 비구매(No) 비율이 상대적으로 높음

- 면적을 통해 범주 간 비율 차이를 시각적으로 파악할 수 있음



<br/><br/><br/>

### 실습 3. 파이차트로 단변량 분석.
purchased 변수 값은 'Yes'(구매함), 'No'(구매 안 함) 두 가지로 나뉩니다.

각 값이 전체 데이터에서 차지하는 비율을 **파이차트**로 시각화해 보세요.

원형 그래프는 전체 중 어느 항목이 많은지 직관적으로 보여줍니다.



```python
# 여기에 코드를 작성하세요!
import matplotlib.pyplot as plt

purchase_counts = df['purchased'].value_counts()

plt.pie(purchase_counts, labels=purchase_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

```python
# 모범 답안
import matplotlib.pyplot as plt


purchase_counts = df['purchased'].value_counts()

plt.pie(purchase_counts, labels=purchase_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Purchase Rate Pie Chart')
plt.axis('equal')
plt.show()
```

##### [결과 해석]
- 'Yes'와 'No' 비율은 같고, 전체 고객 중 실제 구매 비율은 절반이다.

- 원형 면적으로 비율을 직관적으로 이해할 수 있음

- 파이차트는 범주가 2~5개일 때 가장 효과적
