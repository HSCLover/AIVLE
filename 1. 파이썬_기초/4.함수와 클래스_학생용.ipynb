```python

# í´ë˜ìŠ¤ ì •ì˜ ë° ê°ì²´ ìƒì„± ì˜ˆì œ
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"ì•ˆë…•í•˜ì„¸ìš”, ì œ ì´ë¦„ì€ {self.name}ì´ê³ , ë‚˜ì´ëŠ” {self.age}ì‚´ì…ë‹ˆë‹¤.")

# ê°ì²´ ìƒì„±
p1 = Person("ì˜í¬", 25)
p1.introduce()
```
