## 1. ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸° ë° ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
- ë¶“ê½ƒ(iris) ë°ì´í„°ì…‹ì„ ë¶ˆëŸ¬ì˜¤ê³ , Pandas DataFrameìœ¼ë¡œ ë³€í™˜í•˜ëŠ” ë‹¨ê³„ì…ë‹ˆë‹¤.
- ë¶“ê½ƒ(iris) ë°ì´í„°ì…‹ì€ ê½ƒìê³¼ ê½ƒë°›ì¹¨ì˜ ê¸¸ì´/ë„ˆë¹„ë¡œ ì´ë£¨ì–´ì§„ ìˆ˜ì¹˜í˜• ë°ì´í„°ì´ë©°, ì‹¤ì œë¡œëŠ” ë¼ë²¨ì´ ì¡´ì¬í•˜ì§€ë§Œ ì´ë²ˆ ì‹¤ìŠµì—ì„œëŠ” ë¼ë²¨ ì—†ì´ ë¹„ì§€ë„í•™ìŠµìœ¼ë¡œ í´ëŸ¬ìŠ¤í„°ë§ì„ ì§„í–‰í•©ë‹ˆë‹¤.





```python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”!

# 1. ë¼ì´ë¸ŒëŸ¬ë¦¬ import
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# 2. ë¶“ê½ƒ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=['sepal_length','sepal_width','petal_length','petal_width'])
iris_df.head()
```

```python
## ëª¨ë²”ë‹µì•ˆ

# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ import
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
iris = load_iris()
irisDF = pd.DataFrame(data=iris.data, columns=['sepal_length','sepal_width','petal_length','petal_width'])
irisDF.head()
```

<br/><br/>

---

<br/>


## 2. KMeans êµ°ì§‘í™” ìˆ˜í–‰
- í´ëŸ¬ìŠ¤í„° ê°œìˆ˜ë¥¼ 3ìœ¼ë¡œ ì„¤ì •í•˜ì—¬ êµ°ì§‘í™”ë¥¼ ìˆ˜í–‰í•©ë‹ˆë‹¤.
- KMeansëŠ” ëœë¤í•˜ê²Œ ì¤‘ì‹¬ì ì„ ì„ íƒí•œ í›„, ë°˜ë³µì ìœ¼ë¡œ êµ°ì§‘ì„ ì¡°ì •í•˜ë©° ìµœì í™”í•©ë‹ˆë‹¤.

```python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”!
kmeans1 = KMeans(n_clusters=3, random_state=777,max_iter=10000)
kmeans1.fit(iris_df)
```

```python
## ëª¨ë²”ë‹µì•ˆ

kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=777)
kmeans.fit(irisDF)
```

## 3. êµ°ì§‘í™” ê²°ê³¼ í™•ì¸
- ê° ë°ì´í„° í¬ì¸íŠ¸ê°€ ì–´ëŠ êµ°ì§‘(cluster)ì— ì†í–ˆëŠ”ì§€ í™•ì¸í•©ë‹ˆë‹¤.

```python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”!
print(kmeans1.labels_)
```

```python
## ëª¨ë²”ë‹µì•ˆ

print(kmeans.labels_)
```

## 4. ì›ë˜ íƒ€ê²Ÿê°’ê³¼ êµ°ì§‘ ê²°ê³¼ ë¹„êµ
- ì‹¤ì œ íƒ€ê²Ÿ(í’ˆì¢…)ê³¼ êµ°ì§‘ ê²°ê³¼ë¥¼ ë¹„êµí•˜ì—¬ êµ°ì§‘í™”ê°€ ì–´ëŠ ì •ë„ ì˜ë¯¸ ìˆëŠ”ì§€ë¥¼ í™•ì¸í•©ë‹ˆë‹¤.

```python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”!
iris_df['target'] = iris.target
iris_df['cluster'] = kmeans.labels_
iris_df.groupby(['target', 'cluster'])['sepal_length'].count()
```

```python
## ëª¨ë²”ë‹µì•ˆ

irisDF['target'] = iris.target
irisDF['cluster'] = kmeans.labels_
irisDF.groupby(['target', 'cluster'])['sepal_length'].count()
```

## 5. êµ°ì§‘ ì‹œê°í™”


- êµ°ì§‘í™”ê°€ ì˜ ë˜ì—ˆëŠ”ì§€ë¥¼ í™•ì¸í•˜ë ¤ë©´ ë°ì´í„°ë¥¼ 2ì°¨ì› ê³µê°„ì— ì‹œê°í™”í•˜ëŠ” ê²ƒì´ ì¢‹ìŠµë‹ˆë‹¤.
- ê½ƒìì˜ ê¸¸ì´ì™€ ë„ˆë¹„ ê°™ì€ ë‘ ê°œì˜ íŠ¹ì§•ë§Œ ì„ íƒí•´ë„ ë©ë‹ˆë‹¤.
- petal_lengthì™€ petal_width ë‘ ê°œì˜ íŠ¹ì„±ë§Œ ì‚¬ìš©í•˜ì—¬ êµ°ì§‘ ê²°ê³¼ë¥¼ ì‹œê°í™”í•©ë‹ˆë‹¤.



```python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”!
plt.figure(figsize=(6,4))
plt.scatter(iris_df['petal_length'], iris_df['petal_width'], c=iris_df['cluster'], cmap='viridis') # c = êµ°ì§‘ì˜ ì¤‘ì‹¬(center / 'cluster')
plt.colorbar(label='Cluster')
plt.show()
```

```python
## ëª¨ë²”ë‹µì•ˆ

plt.figure(figsize=(6, 4))
plt.scatter(irisDF['petal_length'], irisDF['petal_width'], c=irisDF['cluster'], cmap='viridis')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('KMeans Clustering ')
plt.colorbar(label='Cluster')
plt.show()
```
