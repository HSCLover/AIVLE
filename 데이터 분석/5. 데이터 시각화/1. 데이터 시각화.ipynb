# Marplotlib 기본 사용법

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

* plot() 함수를 사용하여 기본적인 선 그래프를 생성한다.
* marker, linestyle, color 옵션을 이용하여 스타일 지정이 가능하다.
* xlabel(), ylabel(), title()을 이용하여 그래프에 설명을 추가할 수 있다.

위에서 생성한 데이터로 아래 옵션에 따라 `plot` 그래프를 그려보세요.

- 선 색 : 파랑("b")

- X축 라벨 : "X 축"
- Y축 라벨 : "Y 축"

## 히스토그램 (Histogram)

* 데이터를 구간(bins)으로 나누어 빈도수를 막대 형태로 시각화한 그래프입니다.

* 데이터의 **분포 형태(정규분포, 치우침 등)**를 한눈에 파악할 수 있습니다.

* 수치형 변수의 전체적인 분포를 파악할 때 가장 기본적으로 사용하는 도구입니다.

* plt.hist() 또는 sns.histplot()을 사용하여 그릴 수 있습니다.

* 구간 개수(bins)를 조절하면 시각적 해석이 달라질 수 있으므로 주의가 필요합니다.

image.png

```python
data = np.random.randn(1000)  # 정규분포를 따르는 랜덤 데이터 생성
```

위에서 생성한 데이터를 아래 옵션에 따라 히스토그램 그래프로 그려보세요.
- 구간 수(bin) : 30
- 막대 색 : 파랑("b")
- 막대 테두리 색 : 검정("black")
- 막대 투명도 : 0.7
- X축 라벨 : "값"
- Y축 라벨 : "빈도수"
- 그래프 제목 : "히스토그램 예제"

```python
# 여기에 코드를 작성해보세요!
plt.hist(data, bins=30, color="skyblue", edgecolor="black", alpha=0.7)
plt.xlabel("값")
plt.ylabel("빈도수")
plt.title("히스토그램 예제")
plt.show()
```

```python
# 모범답안

import matplotlib.pyplot as plt

plt.hist(data, bins=30, color="blue", edgecolor="black", alpha=0.7)
plt.xlabel("값")
plt.ylabel("빈도수")
plt.title("히스토그램 예제")
plt.show()
```

##  밀도함수 그래프 (KDE Plot)

* 히스토그램보다 연속적인 분포의 흐름을 잘 보여줍니다.

* 확률밀도함수(Probability Density Function)를 추정한 것으로 해석할 수 있습니다.

* seaborn.kdeplot() 또는 sns.histplot(kde=True)로 그릴 수 있습니다.

* 샘플 수가 적거나 이상치가 많을 경우 왜곡될 수 있어 주의해야 합니다.

(비교를 위해 히스토그램 그래프를 그릴때 사용한 데이터를 그대로 활용하겠습니다.)

##  산점도 (Scatter Plot)
* 두 수치형 변수 간의 관계를 점으로 시각화한 그래프입니다.

* 변수 간 상관관계(양의 상관, 음의 상관, 무상관 등)를 확인할 수 있습니다.

* 데이터 분포의 패턴, 이상치, 군집 등을 직관적으로 파악할 수 있습니다.

* plt.scatter() 또는 sns.scatterplot()으로 작성할 수 있습니다.

image.png

데이터는 다음과 같이 만들겠습니다.

산점도 데이터
- X가 0~1 사이의 난수, Y가 $y = 2x + 1 + \text{noise}$를 만족하는 데이터 100개를 생성합니다.  
($\text{noise}$는 평균 0, 표준편차 0.2를 따르는 정규 분포입니다.)  

회귀선 데이터
- X는 0과 1 사이의 값, Y는 $y = 2x + 1$를 만족하는 데이터 100개를 생성합니다.

</br>

위와 같이 만든 데이터를 이용하여 다음 옵션에 맞춰 한 그래프에 산점도와 회귀선을 그려보세요.

산점도 그래프 옵션
- 점 색 : 파랑("b")
- 점 투명도 : 0.5
- 산점도 라벨 : "데이터 포인트"

회귀선 옵션
- 선 색 : 빨강("red")
- 선 종류 : 점선("--")
- 선 라벨 : "회귀선: y = 2x + 1"

그래프 옵션
- X축 라벨 : "변수 X"
- y축 라벨 : "변수 Y"
- 그래프 제목 : "산점도 및 회귀선 예제"

```python
import matplotlib.pyplot as plt

# X 값 생성 (0 ~ 1 사이의 난수 100개)
x = np.random.rand(100)
# Y 값 생성 (y = 2x + 1 + 랜덤 노이즈)
noise = np.random.normal(loc=0, scale=0.2, size=100) # 평균 0, 표준편차 0.2의 정규분포 노이즈 추가
y = 2 * x + 1 + noise

# 여기에 코드를 작성해보세요!

plt.scatter(x,y,color="blue",alpha=0.5, label="데이터 포인트")

xline = np.linspace(0, 1, 100)
yline = 2 * xline + 1
plt.plot(xline, yline, color="orange", linestyle="--", label="회귀선: y = 2x + 1")

plt.xlabel("VAR_X")
plt.ylabel("VAR_Y")
plt.title("산점도 및 회귀선")
plt.legend()
plt.show()
```

```python
# 모범답안

import matplotlib.pyplot as plt

# X 값 생성 (0 ~ 1 사이의 난수 100개)
x = np.random.rand(100)
# Y 값 생성 (y = 2x + 1 + 랜덤 노이즈)
noise = np.random.normal(loc=0, scale=0.2, size=100) # 평균 0, 표준편차 0.2의 정규분포 노이즈 추가
y = 2 * x + 1 + noise

# 산점도 그리기
plt.scatter(x, y, color="blue", alpha=0.5, label="데이터 포인트")

# 회귀선 추가 (y = 2x + 1)
x_line = np.linspace(0, 1, 100) # 0 ~ 1 사이의 100개 점
y_line = 2 * x_line + 1 # y = 2x + 1 함수식
plt.plot(x_line, y_line, color="red", linestyle="--", label="회귀선: y = 2x + 1")

# 그래프 꾸미기
plt.xlabel("변수 X")
plt.ylabel("변수 Y")
plt.title("산점도 및 회귀선 예제")
plt.legend() # 범례 추가
plt.show()
```

## [참고] 페어플롯 (Pair Plot)

* seaborn.pairplot() 함수로 간단하게 구현할 수 있습니다.

* 값이 클수록 진하거나 밝은 색으로 표시되어 한눈에 비교하기 좋습니다.

* sns.heatmap()을 사용하며, annot=True로 수치 표시 가능하게 할 수 있습니다.

* 변수 간 관계 구조를 시각적으로 표현할 때 매우 유용합니다.

image.png

## 모자이크 플롯 (Mosaic Plot)
* 두 개 이상의 범주형 변수 간 관계를 면적 비율로 시각화한 그래프입니다.

* 사각형의 크기는 각 범주 조합의 상대적 빈도수를 나타냅니다.

* 범주형 변수 간의 독립성이나 연관성을 시각적으로 해석할 수 있습니다.

* statsmodels.graphics.mosaicplot을 사용해 생성합니다.

image.png


```python
from statsmodels.graphics.mosaicplot import mosaic

# 예제 데이터 생성 (성별과 구매 여부)
data = {
    '성별': ['남', '여', '남', '여', '남', '여', '남', '남', '여', '여'],
    '구매여부': ['Y', 'N', 'Y', 'Y', 'N', 'N', 'Y', 'N', 'Y', 'N']
}
df = pd.DataFrame(data)
```

위 데이터에서 성별 구매여부에 대해 교차표를 생성하고 이를 모자이크 플롯으로 그려보세요.

```python
# 여기에 코드를 작성해보세요!
```

```python
# 모범답안

import matplotlib.pyplot as plt
import pandas as pd

# 교차표 생성
cross_tab = pd.crosstab(df['성별'], df['구매여부'])

# 모자이크 플롯 그리기
plt.figure(figsize=(6, 5))
mosaic(cross_tab.stack())
plt.title("성별과 구매여부 모자이크 플롯")
plt.show()
```

## 시계열 데이터 시각화
시간에 따른 변화를 나타내는 데이터로, 선 그래프(line chart) 를 주로 사용합니다.

image.png

```python
dates = pd.date_range(start="2024-01-01", periods=100)
values = np.cumsum(np.random.randn(100))  # 랜덤 누적 합계
```

위에서 생성한 시계열 데이터를 아래 옵션에 따라 plot 그래프로 그려보세요.
- 선 종류 : 실선("-")
- 선 색 : 파랑("b")
- 마커 종류 : 원형 마커("o")
- X축 라벨 : "날짜"
- Y축 라벨 : "값"
- 그래프 제목 : "시계열 데이터 시각화"
- X축 눈금 : 45도 회전

```python
# 여기에 코드를 작성해보세요!
```

```python
# 모범답안

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.plot(dates, values, linestyle="-", marker="o", color="blue")
plt.xlabel("날짜")
plt.ylabel("값")
plt.title("시계열 데이터 시각화")
plt.xticks(rotation=45)
plt.show()
```
